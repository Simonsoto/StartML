## Machine Learning Model

Select the prediction variable. As for this case, it is the price of a house, this can be expressed in hundreds of thousands or even more. That is whhy, the first task is to change the prices to thousands.


```python

y = house_melb_df.Price.fillna(house_melb_df.Price.median())
y.isna().any()
```

Select the 'features'. These are the columns that are input into the model. For this case, these are the columns that determine the price of a house. 

```python
#names of the columns of the features that are going to be used
features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']

#Creating the subset of those columns 
X = house_melb_df[features]
X.info()
```

As there are many values that are missing, it was decided to fill this ones with the median of the distribution (this is based on the fact that the median value is more stable against outliers)

```python
# Go through the columns and fill in the missing values
for col in X.columns:
    # calculate the median of the column
    replacement = X[col].median()
    #replace the missing values 
    X[col] = X[col].fillna(value = replacement)
    
```

```python
# first few rows of the features
X.describe()
```

## Building the model

To build the model, we can use the vastly, and well documented library of [scikit-learn](sklearn.org) which contains bunch of methods, models, examples, tests, and in general tools, to make the building of a model super simple. To build a model;
1. **Define** what type of a model will it be? A decision tree? Some other type of model? Some other parameters of the model type are specified too.
2. **Fit** Capture patterns from the provided data. This is the heart of modeling.
3. **Predict** Predict the independent variable from the features.
4. **Evaluate**. Determine how accurate, precise, or the measure you are using, the model's predictions.

```python
# Load the necessary libraries
from sklearn.tree import DecisionTreeRegressor
```

```python
# Define model. Specify the number for random_state to ensure
# same results each run

house_model = DecisionTreeRegressor(random_state=1000)

# Fit the model 
house_model.fit(X,y)
```

Some machine learning models alllow some randomness in the model training and this might cause differences when these are run multiple times. Specifying a number for 'random_state' ensure you get the same results in each run. This is considered a **good practice** 
You can use any value, the model quality won't depend on exactly what value you choose.

Now, there is a fitted model that can be used to make predictions.

In practice, the idea is to use this model to predict the price of the new houses that are coming into the market rather than the houses that already have a price. 

For the sake of testing the fitted model, this is going to be tested in the first 5 observations of the data.|

```python
print('Predictions for the next 5 houses')
houses = X.head()
print(houses)
print('The predictions are')
print(house_model.predict(houses))
```

## Model Validation 
The idea of this part is to evaluate the performance of the model that was created above. In general, almost every model would be evaluated, the important part here is to define how is the model performance going to be measure. (Add reference)

There are many metrics that can be used for summarizing the model quality, a first example is the **Mean Absolute Error** or MAE. As its name describes it, the MAE is calculate by taking the average of the absolute value of the differencer between the prediction and the acutal value. 



```python
#import the metric
from sklearn.metrics import mean_absolute_error

#Create the predictions
predicted_prices = house_model.predict(X)
#Calculate the MAE
mean_absolute_error(y, predicted_prices)
```

What is the problem here? The meausre that was just calculated is an 'in-sample' measure. We use the same sample to build and to evaluate the model. This is not the best. This is a bad way to evaluate as the model was made to 'fit' this data. The model practical value comes from the predictions on new data, data that it has not seen before. This is called **validation/test data.**

Scikit-learn has a function that is called $\text{train_test_split}$

```python
#import the train_test_split 
from sklearn.model_selection import train_test_split

# split the data into trainint and validation data. For both, features
# and targets. The split is based on a random number generator. 
# Supplying a numeric value
train_X, val_X, train_y, val_y = train_test_split(X,y, random_state = 0)

#Define the model 
house_model = DecisionTreeRegressor()

#Fit the model

house_model.fit(train_X, train_y)

## get the predicte prices on the test set
test_values = house_model.predict(val_X)

print(mean_absolute_error(val_y, test_values))
```

## Experimenting with different models.

Now that there is a way to measure the model and the estimations that it does, there is an important part in the job of a data scientist that is to try different methods to see which is the one that brings the best results. It is important that the methods used fit the problem that you are trying to solve. 

The [scikit-learn's documentation](http://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html) contains relevant information that can help create the best possbile model. For example, the Decision Tree model that was created has many options. The most important options are the ones referring to the depth of the tree. (the depth of the tree is how many splits it makes before making a prediciton)

### Underfitting and overfitting.

As the depth increases, the number of splits that the tree divides the data is bigger and this leave every leaf with fewer observations. For example, depth = 2, will create four leaves, while depth = 10 will create 1024 leaves, with few observations on each of those.

That phenomenon is called **overfitting**. This is when the model matches the training data almost perfectly, but it does not do so with the test/validation data, but in case the tree is to shallow, it does not divide the data into distinct groups. This is the case when there are few divisions (imagine 4) and in each group would contain a big variety of houses. The predictions on the training and the test data will be far bad predictions. When this happens (perform bad also in the training data set) it is said that the model is **underfiting**

This can be seen when changing different parameters in the model. 
![title](over_under_fitting.png)

This can be tested with the model that was built before. 


```python
# Function to calculate different Mean average error.

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    '''
    Function to calculate the MAE from a specific model tree regressor
    with a max depth given by the log of the number of leaves.
    '''
    model  = DecisionTreeRegressor(max_leaf_nodes = max_leaf_nodes, random_state = 0)
    model.fit(train_X, train_y)
    
    pred_val = model.predict(val_X)
    
    mae = mean_absolute_error(val_y, pred_val)
    return mae

errors = []

leaf_sizes = [5,10,50,100,250,500,1000,5000,10000]

for leaf in leaf_sizes:
    partial_mae = get_mae(leaf, train_X, val_X, train_y, val_y)
    errors.append(partial_mae)
    print('Max leaves nodes: {} \t\t Mean Absolute Error: {}'.format(leaf, partial_mae))
    
best_leaf = leaf_sizes[errors.index(min(errors))]


print('And the best number of leaves that we can have is {} /n With a Mean Absolute Error of {}'.format(best_leaf, min(errors)))
```

Now that the number of leaves and best tree size that are needed to obtain the best results, using all the data is possible as the results would be the most accurate. In this case, there is no need to keep the validation data as the decisions about the best model have been done. 

```python
#define the final model
final_model = DecisionTreeRegressor(max_leaf_nodes = best_leaf,random_state = 1)

# Predict 
final_model.fit(X, y)

# Made up example 
examples = []
new_house={'Rooms':3,
           'Bathroom':2.5, 
           'Landsize':185.4,
           'Lattitude':-35.8014,
           'Longtitude':146.9958}
examples.append(new_house)
examples = pd.DataFrame(examples)
examples
final_model.predict(examples)
```

## Using more sophisticated Algorithms

**RANDOM FORESTS**

The decision trees used before leave a dificult decision. A deep tree with many leaves will overfit the data as each prediction is coming from historical data from only a few houses in that leaf. On the other hand, a shallow tree with few leaves will perform poorly because it will fail to capture as many distinctions in the raw data. 

All the sophisticated current modelling techniques have this tension between underfitting and overfitting, but there are models that have clever ideas that can lead to a better performance. The best example is the **random forest**.

Random forest uses many trees, and it makes its prediction by averaging the predictions of each component tree (each tree in the forest). It generally has a better predictve accuracy than a single decision tree and it works well with default parameters. With more modelling practice, it is possible to find other models with better performance, but many of those are sensitive to getting the right parameters. 

Building a _random forest_ is similar to how the decision tree was built. The difference is that now `RandomForestRegressor` class is used instead of `DecisionTreeRegressor`

```python
# Import the necessary packages
from sklearn.ensemble import RandomForestRegressor

# Create the RandomForestRegressor object
forest_model = RandomForestRegressor(random_state = 1)

#fit the model
forest_model.fit(train_X, train_y)

#Make the predictions
predictions = forest_model.predict(val_X)

#Calculate the Mean Absolute Error

mae = mean_absolute_error(val_y, predictions)

print('With the Random Forest Regressor, a MAE of {} was achieved'.format(mae))
```

```python
# Comparison of the results 
mae - min(errors)
```

There is a reduction on the mean absolute error. It is about 20.000 in Australian Dollars that is the currency of the houses in the data set. There are still parameters that can be change will creating the Random Forest object similar to the one used on the Decision Tree. But one of the best features of Random Forest is that they generally work reasonably weel even without that tuning.
